```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as ticker

# Assuming your DataFrame is named 'df' and has the columns 'Datetime', 'AEP_MW', and 'New_Time'

# Convert 'New_Time' to total seconds since midnight
df["New_Time_Seconds"] = pd.to_datetime(df["New_Time"], format='%H:%M:%S').dt.time
df["New_Time_Seconds"] = df["New_Time_Seconds"].apply(lambda t: t.hour * 3600 + t.minute * 60 + t.second)

# Create the line plot using Seaborn
sns.lineplot(x="New_Time_Seconds", y="AEP_MW", data=df)

# Customize the plot (optional)
plt.xlabel("Time (Hr:Min:Sec)")
plt.ylabel("AEP_MW")
plt.title("AEP_MW vs. Time")

# Convert seconds back to Hr:Min:Sec format for the x-axis ticks
def format_func(value, tick_number):
    hours = int(value // 3600)
    minutes = int((value % 3600) // 60)
    seconds = int(value % 60)
    return f'{hours:02}:{minutes:02}:{seconds:02}'

# Set the x-axis major formatter
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_func))

# Display the plot
plt.show()
```
